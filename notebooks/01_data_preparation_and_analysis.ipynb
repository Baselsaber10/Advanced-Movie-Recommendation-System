```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Movie Recommendation - Prep") \
    .master("local[*]") \
    .getOrCreate()

print("Spark Started")
```

```python
movies = spark.read.csv("movies.csv", header=True, inferSchema=True)
ratings = spark.read.csv("ratings.csv", header=True, inferSchema=True)

movies.show(5)
ratings.show(5)
```

```python
movies.printSchema()
ratings.printSchema()
```

```python
movies.na.drop().show()
ratings.na.drop().show()
```

```python
movies_clean = movies.dropDuplicates()
ratings_clean = ratings.dropDuplicates()
```

```python
ratings.describe().show()
```

```python
ratings.groupBy("rating").count().orderBy("rating").show()
```

```python
ratings.groupBy("userId").count().orderBy("count", ascending=False).show(10)
```

```python
ratings.groupBy("movieId").count().orderBy("count", ascending=False).show(10)
```

```python
movies_clean.write.mode("overwrite").csv("clean_movies", header=True)
ratings_clean.write.mode("overwrite").csv("clean_ratings", header=True)
```
